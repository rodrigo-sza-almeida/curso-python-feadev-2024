📊 Exercício 1: Análise Dual com Gráficos de Barras e Dispersão

O que você vai fazer:

Crie dois gráficos em uma única visualização:

Um gráfico de barras para mostrar as vendas mensais de uma empresa ao longo de um ano.

Um gráfico de dispersão para mostrar a relação entre as vendas mensais e os meses do ano.

Meses: ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

Vendas: [2100, 2600, 2800, 2300, 2700, 3100, 3300, 3200, 2900, 3000, 3400, 3500]

Adicione:

No gráfico de barras:

Um título: "Vendas Mensais de 2024".
Rótulos no eixo x para os meses e no eixo y para o valor das vendas.
No gráfico de dispersão:

Rótulos para os eixos: "Meses" e "Vendas (R$)".
Um destaque em uma cor única para o mês com a maior venda.
Dica:
Use cores diferentes para os dois gráficos e ajuste as escalas para garantir clareza visual.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
````

````python
# --- Exercício 1 ---

# Dados
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
vendas = [2100, 2600, 2800, 2300, 2700, 3100, 3300, 3200, 2900, 3000, 3400, 3500]

df = pd.DataFrame({'Meses': meses, 'Vendas': vendas})

# Configuração da figura
plt.figure(figsize=(14, 6))

# Gráfico de barras
plt.subplot(1, 2, 1)
plt.bar(df['Meses'], df['Vendas'], color='skyblue')
plt.title('Vendas Mensais de 2024')
plt.xlabel('Meses')
plt.ylabel('Vendas')

# Gráfico de dispersão
plt.subplot(1, 2, 2)
plt.scatter(df['Meses'], df['Vendas'], color='salmon')
plt.title('Vendas Mensais de 2024')
plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.tight_layout()
plt.show()
````

📈 Exercício 2: Gráfico de Barras e Histograma

O que você vai fazer:

Crie dois gráficos em uma única visualização:

Um gráfico de barras que compare as despesas mensais por categoria.

Um histograma que mostre a distribuição das idades dos funcionários de uma empresa.

Categorias: ['Alimentação', 'Transporte', 'Saúde', 'Educação', 'Lazer']

Valores (R$): [1200, 500, 300, 800, 200]

Idades: [23, 25, 22, 30, 35, 40, 29, 28, 45, 50, 24, 27, 32, 33, 48, 55, 25, 26, 31, 37]

Adicione:

No gráfico de barras:

Um título: "Despesas Mensais por Categoria".

Rótulos no eixo x para as categorias e no eixo y para os valores das despesas.

No histograma:

Um título: "Distribuição das Idades dos Funcionários".

Rótulos para os eixos: "Idades" e "Frequência".

````python
# dados
categorias = ['Alimentação', 'Transporte', 'Saúde', 'Educação', 'Lazer']
valores = [1200, 500, 300, 800, 200]
idades = [23, 25, 22, 30, 35, 40, 29, 28, 45, 50, 24, 27, 32, 33, 48, 55, 25, 26, 31, 37]

df = pd.DataFrame({'Categorias': categorias, 'Valores': valores})
series_idades = pd.Series(idades)

# Gráfico de barras
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(df["Categorias"], df["Valores"], color='lightgreen')
plt.title('Despesas Mensais por Categoria')
plt.xlabel('Categorias')
plt.ylabel('Valores (R$)')

# Histograma
plt.subplot(1, 2, 2)
plt.hist(series_idades, bins=10, color='lightcoral', edgecolor='black')
plt.title('Distribuição das Idades dos Funcionários')
plt.xlabel('Idades')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
````

## 🥧 **Exercício 3: Gráfico de Pizza e Gráfico de Linha**

### O que você vai fazer:
1. Crie dois gráficos em uma única visualização:
  - Um **gráfico de pizza** que mostre a participação de mercado de diferentes empresas no setor de tecnologia.
  - Um **gráfico de linha** que mostre a temperatura ao longo da semana.

  ```plaintext
  Empresas: ['Apple', 'Samsung', 'Google', 'Microsoft']
  Participação (%): [40, 30, 20, 10]
  Dias: ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom']
  Temperatura (°C): [22, 24, 19, 23, 25, 20, 21]
  ```

### Adicione:
- No gráfico de pizza:
  - Um título: `"Participação de Mercado no Setor de Tecnologia"`.
  - Rótulos para cada fatia com o nome da empresa e a porcentagem correspondente.

- No gráfico de linha:
  - Um título: `"Temperatura ao Longo da Semana"`.
  - Rótulos no eixo **x** para os dias e no eixo **y** para a temperatura em °C.
  - Destaque o dia com a temperatura mais alta em uma cor diferente.

````python
# Dados
empresas = ['Apple', 'Samsung', 'Google', 'Microsoft']
participacao = [40, 30, 20, 10]
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'Sáb', 'Dom']
temperatura = [22, 24, 19, 23, 25, 20, 21]

df_empresas = pd.DataFrame({'Empresas': empresas, 'Participação': participacao})

df_dias = pd.DataFrame({'Dias': dias, 'Temperatura': temperatura})

# Gráfico de Pizza
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.pie(df_empresas['Participação'], labels=df_empresas["Empresas"], autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'lightcoral'])
plt.title('Participação de Mercado no Setor de Tecnologia')

# Gráfico de Linha
plt.subplot(1, 2, 2)
plt.plot(df_dias['Dias'], df_dias['Temperatura'], marker='o', linestyle='-', color='blue')
plt.title('Temperatura ao Longo da Semana')
plt.xlabel('Dias')
plt.ylabel('Temperatura (°C)')
plt.scatter(dias[temperatura.index(max(temperatura))], max(temperatura), color='red', zorder=5)  # Destaque para o dia com a temperatura mais alta

plt.tight_layout()
plt.show()
````

## 📉 **Exercício 4: Gráfico de Barras e Dispersão com Seaborn**

### O que você vai fazer:

1. Crie dois gráficos em uma única visualização:
    - Um **gráfico de barras** que mostre a média de vendas por categoria de produto.
    - Um **gráfico de dispersão** que mostre a relação entre altura e peso de um grupo de pessoas.

    ```plaintext
     Categorias: ['A', 'B', 'C', 'D']
     Vendas: [100, 150, 300, 250]
     Altura (cm): [150, 160, 170, 180, 190]
     Peso (kg): [50, 60, 70, 80, 90]
    ```

### Adicione:
- No gráfico de barras:
  - Um título: `"Média de Vendas por Categoria"`.
  - Rótulos no eixo **x** para as categorias e no eixo **y** para os valores das vendas.

- No gráfico de dispersão:
  - Rótulos para os eixos: `"Altura (cm)"` e `"Peso (kg)"`.
  - Um destaque em uma cor única para o ponto mais alto e pesado.

````python
# Dados
categorias = ['A', 'B', 'C', 'D']
vendas = [100, 150, 300, 250]
altura = [150, 160, 170, 180, 190]
peso = [50, 60, 70, 80, 90]

df_vendas = pd.DataFrame({'Categorias': categorias, 'Vendas': vendas})

df_pessoas = pd.DataFrame({'Altura': altura, 'Peso': peso})

# Configuração da figura
plt.figure(figsize=(14, 6))

# Gráfico de Barras
plt.subplot(1, 2, 1)
sns.barplot(x='Categorias', y='Vendas', data=df_vendas, hue='Categorias', palette='viridis')
plt.title('Vendas por Categoria')
plt.xlabel('Categorias')
plt.ylabel('Vendas')

# Gráfico de Dispersão
plt.subplot(1, 2, 2)
sns.scatterplot(x="Altura", y="Peso", data=df_pessoas, s=100, color='blue')
plt.title('Relação entre Altura e Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.scatter(df_pessoas.iloc[-1, 0], df_pessoas.iloc[-1, 1], color='red', zorder=5)  # Destaque para o ponto mais alto e pesado

plt.tight_layout()
plt.show()
````

## 🔥 **Exercício 5: Histograma e Heatmap com Seaborn**

### O que você vai fazer:

1. Crie dois gráficos em uma única visualização:
    - Um **histograma** que mostre a distribuição das idades dos participantes de um estudo.
    - Um **heatmap** que mostre a correlação entre as vendas de diferentes produtos.

    ```plaintext
    Idades: [22, 25, 27, 29, 32, 35, 38, 40, 42, 45, 48, 50]
    Produto A: [100, 150, 120, 130, 140, 160, 180]
    Produto B: [80, 90, 110, 120, 115, 130, 140]
    Produto C: [60, 75, 90, 100, 105, 110, 120]
    ```

### Adicione:
- No histograma:
  - Um título: `"Distribuição das Idades dos Participantes"`.
  - Rótulos para os eixos: `"Idades"` e `"Frequência"`.

- No heatmap:
  - Um título: `"Correlação entre Vendas de Diferentes Produtos"`.
  - Rótulos para os eixos: `"Produtos"` e `"Vendas"`.

````python
import pandas as pd

# Dados
idades_estudo = [22, 25, 27, 29, 32, 35, 38, 40, 42, 45, 48, 50]
produto_a = [100, 150, 120, 130, 140, 160, 180]
produto_b = [80, 90, 110, 120, 115, 130, 140]
produto_c = [60, 75, 90, 100, 105, 110, 120]

series_idades = pd.Series(idades_estudo)

vendas_produtos = pd.DataFrame({
    'Produto A': produto_a,
    'Produto B': produto_b,
    'Produto C': produto_c
})

# Configuração da figura
plt.figure(figsize=(14, 6))

# Histograma
plt.subplot(1, 2, 1)
sns.histplot(series_idades, bins=10, color='lightblue', kde=True)
plt.title('Distribuição das Idades dos Participantes')
plt.xlabel('Idades')
plt.ylabel('Frequência')

# Heatmap
correlacao = vendas_produtos.corr()

plt.subplot(1, 2, 2)
sns.heatmap(correlacao, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlação entre Vendas de Diferentes Produtos')
plt.xlabel('Produtos')
plt.ylabel('Vendas')

plt.tight_layout()
plt.show()
````
