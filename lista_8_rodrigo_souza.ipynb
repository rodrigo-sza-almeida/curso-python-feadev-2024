游늵 Exerc칤cio 1: An치lise Dual com Gr치ficos de Barras e Dispers칚o

O que voc칡 vai fazer:

Crie dois gr치ficos em uma 칰nica visualiza칞칚o:

Um gr치fico de barras para mostrar as vendas mensais de uma empresa ao longo de um ano.

Um gr치fico de dispers칚o para mostrar a rela칞칚o entre as vendas mensais e os meses do ano.

Meses: ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

Vendas: [2100, 2600, 2800, 2300, 2700, 3100, 3300, 3200, 2900, 3000, 3400, 3500]

Adicione:

No gr치fico de barras:

Um t칤tulo: "Vendas Mensais de 2024".
R칩tulos no eixo x para os meses e no eixo y para o valor das vendas.
No gr치fico de dispers칚o:

R칩tulos para os eixos: "Meses" e "Vendas (R$)".
Um destaque em uma cor 칰nica para o m칡s com a maior venda.
Dica:
Use cores diferentes para os dois gr치ficos e ajuste as escalas para garantir clareza visual.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
````

````python
# --- Exerc칤cio 1 ---

# Dados
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
vendas = [2100, 2600, 2800, 2300, 2700, 3100, 3300, 3200, 2900, 3000, 3400, 3500]

df = pd.DataFrame({'Meses': meses, 'Vendas': vendas})

# Configura칞칚o da figura
plt.figure(figsize=(14, 6))

# Gr치fico de barras
plt.subplot(1, 2, 1)
plt.bar(df['Meses'], df['Vendas'], color='skyblue')
plt.title('Vendas Mensais de 2024')
plt.xlabel('Meses')
plt.ylabel('Vendas')

# Gr치fico de dispers칚o
plt.subplot(1, 2, 2)
plt.scatter(df['Meses'], df['Vendas'], color='salmon')
plt.title('Vendas Mensais de 2024')
plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.tight_layout()
plt.show()
````

游늳 Exerc칤cio 2: Gr치fico de Barras e Histograma

O que voc칡 vai fazer:

Crie dois gr치ficos em uma 칰nica visualiza칞칚o:

Um gr치fico de barras que compare as despesas mensais por categoria.

Um histograma que mostre a distribui칞칚o das idades dos funcion치rios de uma empresa.

Categorias: ['Alimenta칞칚o', 'Transporte', 'Sa칰de', 'Educa칞칚o', 'Lazer']

Valores (R$): [1200, 500, 300, 800, 200]

Idades: [23, 25, 22, 30, 35, 40, 29, 28, 45, 50, 24, 27, 32, 33, 48, 55, 25, 26, 31, 37]

Adicione:

No gr치fico de barras:

Um t칤tulo: "Despesas Mensais por Categoria".

R칩tulos no eixo x para as categorias e no eixo y para os valores das despesas.

Um t칤tulo: "Distribui칞칚o das Idades dos Funcion치rios".

R칩tulos para os eixos: "Idades" e "Frequ칡ncia".

````python
# dados
categorias = ['Alimenta칞칚o', 'Transporte', 'Sa칰de', 'Educa칞칚o', 'Lazer']
valores = [1200, 500, 300, 800, 200]
idades = [23, 25, 22, 30, 35, 40, 29, 28, 45, 50, 24, 27, 32, 33, 48, 55, 25, 26, 31, 37]

df = pd.DataFrame({'Categorias': categorias, 'Valores': valores})
series_idades = pd.Series(idades)

# Gr치fico de barras
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(df["Categorias"], df["Valores"], color='lightgreen')
plt.title('Despesas Mensais por Categoria')
plt.xlabel('Categorias')
plt.ylabel('Valores (R$)')

# Histograma
plt.subplot(1, 2, 2)
plt.hist(series_idades, bins=10, color='lightcoral', edgecolor='black')
plt.title('Distribui칞칚o das Idades dos Funcion치rios')
plt.xlabel('Idades')
plt.ylabel('Frequ칡ncia')

plt.tight_layout()
plt.show()
````

## 游볺 **Exerc칤cio 3: Gr치fico de Pizza e Gr치fico de Linha**

### O que voc칡 vai fazer:
1. Crie dois gr치ficos em uma 칰nica visualiza칞칚o:
  - Um **gr치fico de pizza** que mostre a participa칞칚o de mercado de diferentes empresas no setor de tecnologia.
  - Um **gr치fico de linha** que mostre a temperatura ao longo da semana.

  ```plaintext
  Empresas: ['Apple', 'Samsung', 'Google', 'Microsoft']
  Participa칞칚o (%): [40, 30, 20, 10]
  Dias: ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'S치b', 'Dom']
  Temperatura (춿C): [22, 24, 19, 23, 25, 20, 21]
  ```

### Adicione:
- No gr치fico de pizza:
  - Um t칤tulo: `"Participa칞칚o de Mercado no Setor de Tecnologia"`.
  - R칩tulos para cada fatia com o nome da empresa e a porcentagem correspondente.

- No gr치fico de linha:
  - Um t칤tulo: `"Temperatura ao Longo da Semana"`.
  - R칩tulos no eixo **x** para os dias e no eixo **y** para a temperatura em 춿C.
  - Destaque o dia com a temperatura mais alta em uma cor diferente.

````python
# Dados
empresas = ['Apple', 'Samsung', 'Google', 'Microsoft']
participacao = [40, 30, 20, 10]
dias = ['Seg', 'Ter', 'Qua', 'Qui', 'Sex', 'S치b', 'Dom']
temperatura = [22, 24, 19, 23, 25, 20, 21]

df_empresas = pd.DataFrame({'Empresas': empresas, 'Participa칞칚o': participacao})

df_dias = pd.DataFrame({'Dias': dias, 'Temperatura': temperatura})

# Gr치fico de Pizza
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.pie(df_empresas['Participa칞칚o'], labels=df_empresas["Empresas"], autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightgreen', 'lightcoral'])
plt.title('Participa칞칚o de Mercado no Setor de Tecnologia')

# Gr치fico de Linha
plt.subplot(1, 2, 2)
plt.plot(df_dias['Dias'], df_dias['Temperatura'], marker='o', linestyle='-', color='blue')
plt.title('Temperatura ao Longo da Semana')
plt.xlabel('Dias')
plt.ylabel('Temperatura (춿C)')
plt.scatter(dias[temperatura.index(max(temperatura))], max(temperatura), color='red', zorder=5)  # Destaque para o dia com a temperatura mais alta

plt.tight_layout()
plt.show()
````

## 游늴 **Exerc칤cio 4: Gr치fico de Barras e Dispers칚o com Seaborn**

### O que voc칡 vai fazer:

1. Crie dois gr치ficos em uma 칰nica visualiza칞칚o:
    - Um **gr치fico de barras** que mostre a m칠dia de vendas por categoria de produto.
    - Um **gr치fico de dispers칚o** que mostre a rela칞칚o entre altura e peso de um grupo de pessoas.

    ```plaintext
     Categorias: ['A', 'B', 'C', 'D']
     Vendas: [100, 150, 300, 250]
     Altura (cm): [150, 160, 170, 180, 190]
     Peso (kg): [50, 60, 70, 80, 90]
    ```

### Adicione:
- No gr치fico de barras:
  - Um t칤tulo: `"M칠dia de Vendas por Categoria"`.
  - R칩tulos no eixo **x** para as categorias e no eixo **y** para os valores das vendas.

- No gr치fico de dispers칚o:
  - R칩tulos para os eixos: `"Altura (cm)"` e `"Peso (kg)"`.
  - Um destaque em uma cor 칰nica para o ponto mais alto e pesado.

````python
# Dados
categorias = ['A', 'B', 'C', 'D']
vendas = [100, 150, 300, 250]
altura = [150, 160, 170, 180, 190]
peso = [50, 60, 70, 80, 90]

df_vendas = pd.DataFrame({'Categorias': categorias, 'Vendas': vendas})

df_pessoas = pd.DataFrame({'Altura': altura, 'Peso': peso})

# Configura칞칚o da figura
plt.figure(figsize=(14, 6))

# Gr치fico de Barras
plt.subplot(1, 2, 1)
sns.barplot(x='Categorias', y='Vendas', data=df_vendas, hue='Categorias', palette='viridis')
plt.title('Vendas por Categoria')
plt.xlabel('Categorias')
plt.ylabel('Vendas')

# Gr치fico de Dispers칚o
plt.subplot(1, 2, 2)
sns.scatterplot(x="Altura", y="Peso", data=df_pessoas, s=100, color='blue')
plt.title('Rela칞칚o entre Altura e Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.scatter(df_pessoas.iloc[-1, 0], df_pessoas.iloc[-1, 1], color='red', zorder=5)  # Destaque para o ponto mais alto e pesado

plt.tight_layout()
plt.show()
````

## 游댠 **Exerc칤cio 5: Histograma e Heatmap com Seaborn**

### O que voc칡 vai fazer:

1. Crie dois gr치ficos em uma 칰nica visualiza칞칚o:
    - Um **histograma** que mostre a distribui칞칚o das idades dos participantes de um estudo.
    - Um **heatmap** que mostre a correla칞칚o entre as vendas de diferentes produtos.

    ```plaintext
    Idades: [22, 25, 27, 29, 32, 35, 38, 40, 42, 45, 48, 50]
    Produto A: [100, 150, 120, 130, 140, 160, 180]
    Produto B: [80, 90, 110, 120, 115, 130, 140]
    Produto C: [60, 75, 90, 100, 105, 110, 120]
    ```

### Adicione:
- No histograma:
  - Um t칤tulo: `"Distribui칞칚o das Idades dos Participantes"`.
  - R칩tulos para os eixos: `"Idades"` e `"Frequ칡ncia"`.

- No heatmap:
  - Um t칤tulo: `"Correla칞칚o entre Vendas de Diferentes Produtos"`.
  - R칩tulos para os eixos: `"Produtos"` e `"Vendas"`.

````python
import pandas as pd

# Dados
idades_estudo = [22, 25, 27, 29, 32, 35, 38, 40, 42, 45, 48, 50]
produto_a = [100, 150, 120, 130, 140, 160, 180]
produto_b = [80, 90, 110, 120, 115, 130, 140]
produto_c = [60, 75, 90, 100, 105, 110, 120]

series_idades = pd.Series(idades_estudo)

vendas_produtos = pd.DataFrame({
    'Produto A': produto_a,
    'Produto B': produto_b,
    'Produto C': produto_c
})

# Configura칞칚o da figura
plt.figure(figsize=(14, 6))

# Histograma
plt.subplot(1, 2, 1)
sns.histplot(series_idades, bins=10, color='lightblue', kde=True)
plt.title('Distribui칞칚o das Idades dos Participantes')
plt.xlabel('Idades')
plt.ylabel('Frequ칡ncia')

# Heatmap
correlacao = vendas_produtos.corr()

plt.subplot(1, 2, 2)
sns.heatmap(correlacao, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correla칞칚o entre Vendas de Diferentes Produtos')
plt.xlabel('Produtos')
plt.ylabel('Vendas')

plt.tight_layout()
plt.show()
````
